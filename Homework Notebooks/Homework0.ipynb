## Part 2 — Cost of Living & Salary (30 points)

### Cost of Living Analysis

This analysis estimates the cost of living for three potential locations after completing the MSBA program: San Francisco, CA; Austin, TX; and Seattle, WA.

### Assumptions:
- Monthly costs are estimated based on average data for each location
- Minimum salary needed covers all expenses with no savings
- Desired salary includes a buffer for savings and lifestyle

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
locations = ['San Francisco, CA', 'Austin, TX', 'San Diego, CA']

monthly_costs = {
    'San Francisco, CA': {
        'housing': 2500,  
        'food': 600,
        'transportation': 200,
        'utilities': 150,  
        'entertainment': 300
    },
    'Austin, TX': {
        'housing': 1500,
        'food': 500,
        'transportation': 150,
        'utilities': 100,
        'entertainment': 250
    },
    'San Diego, CA': {
        'housing': 2200,
        'food': 600,
        'transportation': 180,
        'utilities': 120,
        'entertainment': 350
    }
}

desired_monthly_savings = 1000  
desired_lifestyle_buffer = 500   
```

```python
df_costs = pd.DataFrame(monthly_costs).T
df_costs['total_monthly'] = df_costs.sum(axis=1)

print("Monthly Cost of Living by Category:")
df_costs
```

```python

df_costs['min_annual_salary'] = df_costs['total_monthly'] * 12

df_costs['desired_annual_salary'] = (df_costs['total_monthly'] + desired_monthly_savings + desired_lifestyle_buffer) * 12

print("\nSalary Requirements:")
salary_df = df_costs[['total_monthly', 'min_annual_salary', 'desired_annual_salary']]
salary_df
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(locations, df_costs['total_monthly'], color=['blue', 'green', 'red'])
plt.title('Total Monthly Cost of Living by Location')
plt.xlabel('Location')
plt.ylabel('Monthly Cost (USD)')
plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 50,
             f'${height:,.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
