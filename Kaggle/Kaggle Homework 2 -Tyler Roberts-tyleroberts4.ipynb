4. How did boosting versus bagging compare for your work? 

Comparing boosting and bagging in this project was really interesting because both approaches performed well, but in slightly different ways. The Random Forest (bagging) model worked by building many independent trees and averaging their results, which made it very stable and strong overall. It had excellent precision and very high overall accuracy, but it struggled a bit more with capturing the harder cases, like the “High” irrigation class, where recall was lower.

On the other hand, XGBoost (boosting) builds trees sequentially, where each new tree tries to fix the mistakes of the previous ones. This gave it an advantage in capturing more nuanced patterns in the data. You can see this in the slightly higher recall for the harder class and more balanced performance across categories. Even though the overall accuracy was similar, boosting did a better job of handling edge cases, which is often more important in real-world applications.

In my opinion, boosting was slightly better for this problem because it focused more on correcting errors, while bagging focused more on reducing variance. The difference wasn’t huge, but it was consistent. This reinforces the idea that for structured/tabular data, boosting methods like XGBoost are often the best choice when you want to squeeze out the last bit of performance.
